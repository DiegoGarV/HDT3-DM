### Normalidad

```python
# Prueba Shapiro-Wilk
stat, p = shapiro(df['SalePrice'])
print(f'Estadístico: {stat}, p-valor: {p}')

if p > 0.05:
    print("Los datos parecen seguir una distribución normal")
else:
    print("Los datos NO siguen una distribución normal")
```

```python
# Prueba Kolmogorov-Smirnov
stat, p = kstest(df['SalePrice'], 'norm')
print(f'Estadístico: {stat}, p-valor: {p}')
```

```python
# Prueba Anderson-Darling
result = anderson(df['SalePrice'])
print('Estadístico:', result.statistic)
print('Valores críticos:', result.critical_values)
```

```python
# Histograma
sns.histplot(df['SalePrice'], kde=True)
plt.show()

# Q-Q Plot
stats.probplot(df['SalePrice'], dist="norm", plot=plt)
plt.show()
```

Los datos claramente no siguen una distribución normal y es claro en todas las pruebas. El histograma también muestra un sesgo claro y el q-q plot nos muestra que los datos siguen una distribución asimétrica y no lineal. Haremos el analisis de los datos individuales con esto en mente.
