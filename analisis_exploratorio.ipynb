### Normalidad

```python
# Prueba Shapiro-Wilk
stat, p = shapiro(df['SalePrice'])
print(f'Estadístico: {stat}, p-valor: {p}')

if p > 0.05:
    print("Los datos parecen seguir una distribución normal")
else:
    print("Los datos NO siguen una distribución normal")
```

```python
# Prueba Kolmogorov-Smirnov
stat, p = kstest(df['SalePrice'], 'norm')
print(f'Estadístico: {stat}, p-valor: {p}')
```

```python
# Prueba Anderson-Darling
result = anderson(df['SalePrice'])
print('Estadístico:', result.statistic)
print('Valores críticos:', result.critical_values)
```

```python
# Histograma
sns.histplot(df['SalePrice'], kde=True)
plt.show()

# Q-Q Plot
stats.probplot(df['SalePrice'], dist="norm", plot=plt)
plt.show()
```

Los datos claramente no siguen una distribución normal y es claro en todas las pruebas. El histograma también muestra un sesgo claro y el q-q plot nos muestra que los datos siguen una distribución asimétrica y no lineal. Haremos el analisis de los datos individuales con esto en mente.

```python
# Lista de características especiales
features = ['PoolArea', 'Fence', 'MiscVal', 'GarageType', 'FireplaceQu']

# Crear subgráficos
fig, axes = plt.subplots(len(features), 1, figsize=(8, len(features) * 4))

for i, feature in enumerate(features):
    if feature in df.columns:
        sns.boxplot(x=df[feature], y=df['SalePrice'], ax=axes[i])
        axes[i].set_title(f'Relación entre {feature} y el Precio de Venta')

plt.tight_layout()
plt.show()

# Convertir variables categóricas a valores numéricos para calcular la correlación
df_encoded = df.copy()
for feature in features:
    if feature in df_encoded.columns:
        df_encoded[feature] = df_encoded[feature].astype('category').cat.codes

correlations = {feature: df_encoded[feature].corr(df_encoded['SalePrice'], method='spearman') for feature in features if feature in df_encoded.columns}

print("Correlaciones con el precio de venta:")
for feature, corr in correlations.items():
    print(f"{feature}: {corr:.4f}")
```

En este caso podemos ver como las características extra que puede tener una vivienda afectan en el precio de venta de la misma. Lo primero que notamos es que las correlaciones parecen ser muy débiles, incluso llegando a ser negativas. Sin embargo, esto puede deberse a que las categorías están cargadas de datos atípicos, como lo podemos ver en los gráficos de caja y bigote. Analizando estos mismos gráficos podemos ver que ninguna característica parece influir de manera constante en el precio de una vivienda, al menos no de manera significativa, ya que el rango de precios no se ve afectado. La única característica que si llega a influenciar en el precio es la de la piscina, ya que según su tamaño si pueden llegar a volverse bastante caras las casas, pero ocurre en tan pocos casos que puede que no sea el único factor que afecte realmente. 

```python
# Gráfico de caja para visualizar atíipicos
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['SalePrice'])
plt.title('Boxplot de los Precios de Venta')
plt.show()

# Cantidad de atipicos exacta
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
atipicos = df[(df['SalePrice'] < limite_inferior) | (df['SalePrice'] > limite_superior)]

print(f"Número de casas vendidas atípicas: {len(atipicos)}")
```

```python
# Ver estadísticas de las casas vendidas atípicas
print(atipicos.describe())

# Comparar características clave con el resto de las casas
main_chars = ['SalePrice', 'GrLivArea', 'LotArea', 'LotFrontage', 'YearBuilt']
df_comparacion = df[main_chars].copy()
df_comparacion['Outlier'] = df['SalePrice'].apply(lambda x: 'Atípica' if x < limite_inferior or x > limite_superior else 'Normal')

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_comparacion, x='Outlier', y='SalePrice')
plt.title('Comparación de Precios entre Casas Normales y atipicos')
plt.show()

```

Aquí podemos ver que características tienen aquellas casas con valores atípicos al momento de venderlas. Cabe resaltar que estos valores atípicos siempre están por arriba de lo que se espera del valor común de una casa y no por debajo. Con el describe podemos ver que no todas las casas extremadamente caras lo son por la misma razón. Algunas son muy grandes, otras tienen garajes o sotanos más grandes, una mejor condición, remodelaciones actuales, etc. Es importante saber esto para no tomar el tamaño de una casa como el su única fuente de valor, como es común pensar. El gráfico de caja que le sigue es para que se pueda ver en donde se mantiene la media de las casas con precios más "comunes" a comparación de aquellas con precios elevados.
